````python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load the penguins dataset from the provided URL
penguins_url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(penguins_url)

# Create a subplot for each species
species = penguins['species'].unique()
fig = make_subplots(rows=1, cols=len(species), subplot_titles=species, shared_yaxes=True)

# Colors for each statistical marker
colors = {
    'mean': 'blue',
    'median': 'green',
    'range': 'lightgrey',
    'iqr': 'yellow',
    'std_dev': 'red'
}

# Loop over each species and plot histograms with statistical annotations
for i, sp in enumerate(species):
    # Filter data for the species
    data = penguins[penguins['species'] == sp]['flipper_length_mm'].dropna()
    
    # Calculate statistics
    mean = data.mean()
    median = data.median()
    minimum = data.min()
    maximum = data.max()
    std_dev = data.std()
    lower_q, upper_q = data.quantile(0.25), data.quantile(0.75)
    
    # Add histogram for the species
    fig.add_trace(
        go.Histogram(x=data, nbinsx=20, name=f'{sp} Flipper Length', marker_color='lightblue', showlegend=False),
        row=1, col=i+1
    )
    
    # Add range (min to max) as a shaded area
    fig.add_vrect(x0=minimum, x1=maximum, fillcolor=colors['range'], opacity=0.2, line_width=0, row=1, col=i+1)
    
    # Add interquartile range (IQR) as a shaded area
    fig.add_vrect(x0=lower_q, x1=upper_q, fillcolor=colors['iqr'], opacity=0.3, line_width=0, row=1, col=i+1)
    
    # Add ±2 standard deviations as a shaded area
    fig.add_vrect(x0=mean - 2*std_dev, x1=mean + 2*std_dev, fillcolor=colors['std_dev'], opacity=0.1, line_width=0, row=1, col=i+1)
    
    # Add mean and median lines
    fig.add_vline(x=mean, line=dict(color=colors['mean'], dash='dash'), row=1, col=i+1, annotation_text="Mean", annotation_position="top left")
    fig.add_vline(x=median, line=dict(color=colors['median'], dash='dot'), row=1, col=i+1, annotation_text="Median", annotation_position="top right")

# Update layout for better readability
fig.update_layout(
    title_text='Flipper Length Distributions with Statistical Annotations for Each Species',
    height=600, width=1200,
    showlegend=False,
    xaxis_title='Flipper Length (mm)',
    yaxis_title='Count',
    margin=dict(t=50, l=50, r=50, b=50)  # Adjust margins
)

# Display the figure
fig.show(renderer="png")
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the penguins dataset
penguins_url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(penguins_url)


# Define colors for the statistical markers
colors = {
    'mean': 'blue',
    'median': 'green',
    'range': 'lightgrey',
    'iqr': 'yellow',
    'std_dev': 'red'
}

# Create a figure with subplots (1 row, 3 columns) for each species
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
species = penguins['species'].unique()

# Loop over each species and create a KDE plot
for i, sp in enumerate(species):
    ax = axes[i]
    
    # Filter the data for the current species
    data = penguins[penguins['species'] == sp]['flipper_length_mm'].dropna()
    
    # Plot the KDE (kernel density estimation) for flipper_length_mm
    sns.kdeplot(data, ax=ax, color='lightblue', fill=True, alpha=0.6, label=f'{sp} KDE')
    
    # Calculate statistical values
    mean = data.mean()
    median = data.median()
    minimum = data.min()
    maximum = data.max()
    std_dev = data.std()
    lower_q, upper_q = data.quantile(0.25), data.quantile(0.75)
    
    # Add vertical lines and shaded areas for statistical markers
    ax.axvline(mean, color=colors['mean'], linestyle='--', label='Mean')
    ax.axvline(median, color=colors['median'], linestyle=':', label='Median')
    
    # Shade the range (min to max)
    ax.axvspan(minimum, maximum, color=colors['range'], alpha=0.2, label='Range (min-max)')
    
    # Shade the IQR (interquartile range)
    ax.axvspan(lower_q, upper_q, color=colors['iqr'], alpha=0.3, label='Interquartile Range (IQR)')
    
    # Shade the ±2 standard deviations
    ax.axvspan(mean - 2*std_dev, mean + 2*std_dev, color=colors['std_dev'], alpha=0.1, label='±2 Std. Deviations')
    
    # Set title and labels
    ax.set_title(f'{sp} Flipper Length KDE')
    ax.set_xlabel('Flipper Length (mm)')
    ax.set_ylabel('Density')

# Add a shared legend outside the plot
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=5, bbox_to_anchor=(0.5, 1.1), fontsize='medium')

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Display the plot
plt.show()
 
````

1. Box Plot
Pros:
Compact summary: Displays key summary statistics (minimum, Q1, median, Q3, maximum) in a single, easy-to-read visual.
Outlier detection: Effectively identifies and highlights outliers.
Comparison: Easy to compare distributions across different groups or categories.
Simple to understand: Straightforward for audiences who understand basic statistics.
Cons:
Lack of detailed shape: Does not show the internal distribution shape (e.g., skewness, modality).
No frequency information: Does not provide information about the frequency of data points.
Data hiding: Two datasets with the same five-number summary (min, Q1, median, Q3, max) may look identical in a box plot, even if their distributions are very different.
2. Histogram
Pros:
Intuitive: Simple to create and easy to understand; visually shows data distribution by frequency.
Reveals distribution shape: Highlights the shape of the distribution (e.g., skewness, modality).
Flexible bin width: Bins can be adjusted to reveal finer or coarser details about the distribution.
Cons:
Bin sensitivity: The choice of bin width can drastically change the perceived distribution, leading to possible misinterpretation.
Not compact: It can be harder to compare multiple histograms for different groups side by side.
No summary statistics: Does not explicitly provide summary statistics like median or quartiles without additional computation.
Discrete nature: Bins give a segmented view that may fail to capture subtleties of continuous data.
3. Kernel Density Estimator (KDE)
Pros:
Smooth visualization: Provides a continuous and smooth estimate of the data distribution, avoiding the binning issues of histograms.
Better for small datasets: With appropriate bandwidth selection, KDE can show distribution shape more effectively than histograms with small sample sizes.
Visual clarity: Often more visually appealing and clear in representing overall distribution trends.
Flexible bandwidth: Allows control over the degree of smoothness applied to the distribution, enabling detailed analysis when needed.
Cons:
Over-smoothing: If the bandwidth is too large, the KDE can over-smooth the data and hide important details like peaks or outliers.
Less intuitive: May be harder for non-technical audiences to interpret compared to histograms and box plots.
Computational cost: More computationally intensive than histograms, especially with large datasets.
No summary statistics: Does not provide explicit summary statistics such as quartiles, median, or range.


My preference:I prefer histograms because they offer a clear, straightforward way to visualize the frequency distribution of data, making it easy to understand how data is spread across different ranges. Thus, histograms are a great way to show results of numeric data, such as weight, height, time, etc.
Unlike box plots, histograms provide more detailed insights into the density of data within specific intervals, which is useful for detecting patterns like unimodal, bimodal, or multimodal distributions. 
The ability to adjust bin sizes in histograms allows me to focus on specific levels of detail, giving flexibility based on the level of granularity I need. For example, smaller bins provide finer detail, while larger bins offer a more generalized view of the distribution.
Histograms are also intuitive and familiar to many people, making it easier for audience to interpret and understand the data.

Skewness occurs when data isn't symmetrically distributed around the center. 

Right Skewness (Positive Skewness): In this case, the tail on the right side (toward larger values) of the distribution is longer or fatter. The mean is typically greater than the median because the long tail on the right pulls the mean higher.

Left Skewness (Negative Skewness): In this case, the tail on the left side (toward smaller values) is longer or fatter. The mean tends to be smaller than the median because outliers or smaller values pull the mean toward the left.
We can compare the properties of a right-skewed distribution with a left-skewed distribution by calculating their means and medians, and visualizing their histograms.

````python
#Right-Skewed Distribution Example (Positive Skew)

import pandas as pd
import numpy as np
import plotly.express as px
from scipy import stats

# Right-skewed distribution: Gamma Distribution
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

# Visualizing the histogram
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Right-Skewed Distribution (Positive Skew)")
fig1.show(renderer="png")  # for GitHub/MarkUs submission

# Mean and Median of the right-skewed distribution
mean1 = sample1.mean()
median1 = np.median(sample1)

mean1, median1
````

````python
#Left-Skewed Distribution Example (Negative Skew)
sample2 = -stats.gamma(a=2, scale=2).rvs(size=1000)

# Visualizing the histogram
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Left-Skewed Distribution (Negative Skew)")
fig2.show(renderer="png")  # for GitHub/MarkUs submission

# Mean and Median of the left-skewed distribution
mean2 = sample2.mean()
median2 = np.median(sample2)

mean2, median2
````

For the right-skewed distribution, the mean is larger than the median because of the few larger values (outliers) that pull the mean to the right.

For the left-skewed distribution, the mean will be smaller than the median, as negative outliers pull the mean to the left, making the median larger than the mean.

````python
import pandas as pd
import plotly.express as px

# Load the dataset
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')

# Make identical boy and girl names distinct by appending sex to the name
bn['name'] = bn['name'] + " " + bn['sex']

# Rank names by percent in each year
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)
bn = bn.sort_values(['name', 'year'])

# Create the 'percent change' column to track the increase or decrease in name prevalence from the last year
bn['percent change'] = bn['percent'].diff()

# Identify new names and set percent change for the first occurrence
new_name = [True] + list(bn.name[:-1].values != bn.name[1:].values)
bn.loc[new_name, 'percent change'] = bn.loc[new_name, 'percent']

# Sort by year and restrict to "common" names with percent > 0.001
bn = bn.sort_values('year')
bn = bn[bn.percent > 0.001]

# Create the scatter plot with the specified parameters
fig = px.scatter(bn, 
                 x="percent change",         
                 y="rank",                    
                 animation_frame="year",     
                 animation_group="name",     
                 size="percent",            
                 color="sex",              
                 hover_name="name",         
                 size_max=50,                
                 range_x=[-0.005, 0.005])     

# Reverse the y-axis to put rank 1 at the top
fig.update_yaxes(autorange='reversed')

# Show the plot
fig.show(renderer="png")
````
