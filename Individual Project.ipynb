Analysis 2: Confidence Intervals
1. Research Question
Research Question: Is there a significant difference in the level of agreeableness(measured by the Big Five Inventory)among different gender groups in the surveyed population?
Relevance: This question aims to explore potential gender-based differences in the agreeableness trait. Understanding the differences could provide valuable insights into behavioral tendencies associated with gender identity and may help in tailoring interventions or interpreting other psychological dimensions in the dataset.
2. Variables to Explore
Agreeableness Score (PSYCH_big_five_inventory_agreeable_score): This score captures the agreeableness trait, a key dimension in personality psychology.

Gender: The analysis will focus on different gender identities, including categories such as "Man," "Woman," and "Non-binary" (with subcategories including Two Spirit, agender, genderfluid, genderqueer, or other terms).
Justification for Variables:

Agreeableness is relevant because it provides a quantitative measure of a personality trait that might vary by gender identity. Gender grouping allows us to examine whether there is difference in agreeableness scores among genders.

Visualization:
Box Plot to display the distribution of agreeableness scores within each gender group, highlighting median, range, and outliers.
Use a normal distribution curve centered on the mean agreeableness score for each gender, with shaded regions representing the confidence interval.
3. Planned Analysis
Method:
Confidence Intervals for Means: I will calculate 95% confidence intervals for the mean agreeableness score within each gender group to understand the degree of agreement in each group.
Bootstrapping: To construct these intervals, I will use bootstrapping to generate confidence intervals by resampling the data. This method is particularly useful if the sample sizes for certain gender categories are small or the data distribution is non-normal.
Assumptions:
We assume that the bootstrapping method provides a reliable estimate of the confidence intervals, especially if the gender group sample sizes are relatively small.
The analysis assumes that the sample data for each gender group is representative of the larger population for which we are trying to make inferences.
4. Hypothesis and Expected Results

Null Hypothesis: There is no significant difference in the mean agreeableness scores between gender groups; any observed difference is due to random sampling variation.
Alternative Hypothesis: There is a significant difference in the mean agreeableness scores between gender groups.
Expected Results:

If the confidence intervals for the mean agreeableness scores of the gender groups do not overlap, it would indicate a statistically significant difference between groups. For example, if "Man" and "Woman" groups have non-overlapping intervals, it suggests a real difference in agreeableness scores.
If the confidence intervals do overlap, it would suggest no significant difference in the means, implying similar levels of agreeableness across gender identities.
Interpretation and Relevance:

Finding distinct mean differences between gender groups could inform targeted approaches in mental health or personality studies. For example, if one gender group consistently scores higher on agreeableness, programs focusing on social harmony and interpersonal relationships could be tailored toward that group.
Overlapping intervals would suggest a general consistency in agreeableness scores across gender identities, reinforcing the notion of similarity rather than difference in this personality trait across gender categories.
 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv"
data = pd.read_csv(url)

# Data Preparation
data = data[['DEMO_gender', 'PSYCH_big_five_inventory_agreeable_score']].dropna()
data['DEMO_gender'] = data['DEMO_gender'].astype('category')

# Summary Statistics
summary_stats = data.groupby('DEMO_gender')['PSYCH_big_five_inventory_agreeable_score'].agg(['mean', 'std', 'count'])
print("Summary Statistics by Gender:")
print(summary_stats)

# Calculate 95% Confidence Intervals
confidence_intervals = {}
for gender, row in summary_stats.iterrows():
    mean = row['mean']
    std_dev = row['std']
    n = row['count']
    std_err = std_dev / np.sqrt(n)
    ci = 1.96 * std_err  # 95% CI margin
    confidence_intervals[gender] = (mean - ci, mean + ci)
print("95% Confidence Intervals by Gender:")
print(confidence_intervals)

# Visualization: Normal Distribution with Confidence Intervals
plt.figure(figsize=(12, 6))
x_values = np.linspace(20, 40, 200)  # Adjust range based on expected agreeableness score range

for gender, row in summary_stats.iterrows():
    mean = row['mean']
    std_dev = row['std']
    n = row['count']
    
    # Plot the normal distribution curve
    y_values = norm.pdf(x_values, mean, std_dev / np.sqrt(n))
    plt.plot(x_values, y_values, label=f"{gender} (Mean: {mean:.2f})")
    
    # Shade the confidence interval region
    ci_lower, ci_upper = confidence_intervals[gender]
    x_ci = np.linspace(ci_lower, ci_upper, 100)
    y_ci = norm.pdf(x_ci, mean, std_dev / np.sqrt(n))
    plt.fill_between(x_ci, y_ci, alpha=0.3)

# Add labels and legend
plt.xlabel('Agreeableness Score')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curves with 95% Confidence Intervals by Gender')
plt.legend()
plt.show()
```

5. Analysis Execution

Woman group has the highest mean agreeableness score (32.68), with a narrow confidence interval (32.41, 32.94), making this estimate very reliable due to the large sample size.
Man group has a mean score of 31.20, with a confidence interval (30.63, 31.78), indicating some overlap with the Woman group, suggesting similar agreeableness levels.
Non-binary (with subcategories) group has the lowest mean score (29.59), with a confidence interval (28.38, 30.80), indicating less overlap with the Woman group and suggesting a potential difference in agreeableness levels.
Non-binary group (small sample) has a mean score of 31.33, but the wide confidence interval (27.52, 35.14) means this estimate is less certain.
The groups have relatively similar means, mostly within a few units, suggesting some consistency across the population.

Analysis 3: Simple Linear Regression

1. Research Question
Research Question: Is there a relationship between age and life satisfaction among the surveyed population?

Relevance: Understanding how life satisfaction changes with age can reveal valuable insights into the well-being of individuals across different ages. This information could inform policies and programs aimed at enhancing quality of life for specific age groups.
2. Variables to Explore
Predictor Variable:
Age (DEMO_age): This variable represents the age of the participants, which will serve as the predictor in the regression model.

Response Variable:
Life Satisfaction Score (WELLNESS_life_satisfaction): This variable measures self-reported life satisfaction, which we aim to predict based on age.

Age is an essential demographic factor that could influence well-being, while life satisfaction is a key indicator of overall happiness and quality of life.
Exploring these variables helps identify trends in well-being across the lifespan.

3. Planned Analysis
Method:

Simple Linear Regression: We will perform a simple linear regression to assess the relationship between age (predictor) and life satisfaction (response). This involves estimating the regression line, calculating the slope and intercept, and interpreting the strength and direction of the relationship.
Goodness of Fit (R² Value): We will calculate the R² value to understand how well age explains the variation in life satisfaction scores. A higher R² would suggest that age is a significant predictor of life satisfaction.

Assumptions:
Linearity: We assume a linear relationship between age and life satisfaction.
Independence: The observations should be independent of each other.
Homoscedasticity: We assume constant variance of life satisfaction scores across different ages.
Normality of Residuals: We assume that the residuals (differences between observed and predicted values) are normally distributed.
4. Hypothesis and Expected Results
Hypothesis:
Null Hypothesis: There is no significant relationship between age and life satisfaction.
Alternative Hypothesis: There is a significant relationship between age and life satisfaction.
Possible Results:

Positive Slope: A positive slope would suggest that life satisfaction tends to increase with age.
Negative Slope: A negative slope would indicate that life satisfaction decreases as age increases.
No Significant Relationship: If the slope is near zero and the p-value is not significant, it would suggest that age does not substantially impact life satisfaction.

Interpretation and Relevance:
If there is a positive or negative trend, it would suggest that life satisfaction changes predictably with age, providing insights into specific life stages where satisfaction may peak or dip.
If no significant relationship is found, it suggests that factors other than age are more relevant for life satisfaction, highlighting the need to consider additional variables in well-being studies.
 

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

# Load the CSCS dataset
data_url = 'https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv'
data = pd.read_csv(data_url)

# Check available columns
print(data.columns)

# Handle missing values by dropping rows with NaN values in the selected columns
data = data[['DEMO_age', 'WELLNESS_life_satisfaction']].dropna()

# Define the variables
X = data['DEMO_age']  # Independent variable: Age
Y = data['WELLNESS_life_satisfaction']  # Dependent variable: Life Satisfaction

# Add a constant to the independent variable
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(Y, X).fit()

# Print the summary of the regression
print(model.summary())

# Plot the data points and regression line
plt.scatter(data['DEMO_age'], data['WELLNESS_life_satisfaction'], color='blue', label='Data Points')
plt.plot(data['DEMO_age'], model.predict(X), color='red', label='Regression Line')
plt.xlabel('Age')
plt.ylabel('Life Satisfaction Score')
plt.title('Simple Linear Regression: Age vs. Life Satisfaction')
plt.legend()
plt.show()
```

The regression line shows a slight positive slope, indicating a weak positive relationship between age and life satisfaction. This suggests that as age increases, there is a slight increase in life satisfaction, although the effect is minimal.
Goodness of Fit (R² Value):

The weak positive trend suggests a slight increase in life satisfaction with age, but the relationship is not strong. The wide spread of data points at each age level indicates that age alone cannot fully explain life satisfaction.
3.6 Relevance of Analysis
Implications for Well-Being Programs:
Understanding whether age significantly influences life satisfaction can help guide support programs for different age groups. For example, if life satisfaction drops at certain ages, targeted support could help improve well-being during these stages.
The weak relationship here suggests that while age may have some impact, other factors should also be considered when designing programs to enhance well-being. This aligns with the project’s objective of examining well-being factors.
