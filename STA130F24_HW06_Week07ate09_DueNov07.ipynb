The Simple Linear Regression model assumes that there is a linear relationship between a predictor (independent) variable x and an outcome (dependent) variable ùëå. The model is defined by the equation:

ùëåùëñ=ùõΩ0+ùõΩ1xi+œµi
Yi represents the outcome variable for observation 
xi represents the predictor variable for observation  
Œ≤0 is a constant term representing the expected value of ùëå when x=0
Œ≤1(slope) represents the change in ùëå for a one-unit increase in ùë•
œµi is the error term or "noise" for each observation, representing the variation in ùëå that cannot be explained by the linear relationship alone.

In this model, each outcome Yi is generated as the sum of a deterministic part  (Œ≤0+Œ≤1xi)and a random component (œµi).Since œµi follows a normal distribution,Yi also follows a normal distribution centered around the value Œ≤0+Œ≤1xi and with a standard deviation of œÉ.Thus, for each fixed ùë•ùëñ, the corresponding Yi ,values form a sample from a normal distribution around the regression line Œ≤0+Œ≤1xi.


````python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import norm

# Define parameters for the model
n = 100  # number of data points
beta0 = 2  # intercept
beta1 = 3  # slope
sigma = 1.5  # standard deviation of the error term

# Generate x values
x = np.linspace(0, 10, n)

# Generate random error terms from a normal distribution
errors = norm.rvs(loc=0, scale=sigma, size=n)

# Calculate theoretical Y values without errors (for the line)
Y_theoretical = beta0 + beta1 * x

# Calculate actual Y values using the theoretical model with errors
Y = Y_theoretical + errors

# Create a Plotly figure
fig = go.Figure()

# Add the theoretical line (without errors)
fig.add_trace(go.Scatter(x=x, y=Y_theoretical, mode='lines', name='Theoretical Line (a + bx)', line=dict(dash='dash')))

# Add the simulated data points (with errors)
fig.add_trace(go.Scatter(x=x, y=Y, mode='markers', name='Simulated Data Points'))

# Customize the layout
fig.update_layout(
    title='Visualization of Theoretical Line and Simulated Data Points',
    xaxis_title='x',
    yaxis_title='Y',
    showlegend=True
)

# Show the plot
fig.show(renderer="png")
````

````python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf  # for statistical modeling, including OLS regression
import plotly.express as px  # for creating interactive plots

# Simulate data for x and Y
np.random.seed(0)  # for reproducibility
x = np.random.rand(100) * 10  # example independent variable data
Y = 2.5 * x + np.random.normal(size=100)  # example dependent variable data with some noise

# Combine x and Y into a DataFrame named df
df = pd.DataFrame({"x": x, "Y": Y})

# Specifying and fitting the OLS model
model_data_specification = smf.ols("Y~x", data=df)  # define Y as dependent and x as independent
fitted_model = model_data_specification.fit()  # fit model to data and store results

# Getting information from the fitted model
fitted_model.summary()  # summary of the model, including fit quality and coefficients
fitted_model.summary().tables[1]  # table with parameter estimates (intercept and slope)
fitted_model.params  # estimated values of the model parameters
fitted_model.params.values  # numerical values of the parameters (without labels)
fitted_model.rsquared  # R-squared value, shows how well the model fits the data

# Plotting data and adding the OLS trendline
df['Data'] = 'Data'  # label data to show in the plot legend
fig = px.scatter(df, x='x',  y='Y', color='Data', trendline='ols', title='Y vs. x')  # scatter plot with trendline

# Manually adding the OLS trendline
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues, line=dict(color='blue'), name="trendline='ols'")  # custom trendline

fig.show(renderer="png")  # use fig.show(renderer="png") for static images on GitHub and MarkUs
````

> 
> _$$\hat \beta_1 = r_{xy}\frac{s_y}{s_x} \quad \text{ and } \quad  \hat\beta_0 = \bar {y}-\hat \beta_1\bar {x}$$_

> _where $r_{xy}$ is the **correlation** between $x$ and $Y$ and $s_x$ and $s_Y$ are the **sample standard deviations** of $x$ and $y$_

````python
import scipy.stats as stats
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px

n,x_min,x_range,beta0,beta1,sigma = 20,5,5,2,3,5
x = stats.uniform(x_min, x_range).rvs(size=n)
errors = stats.norm(loc=0, scale=sigma).rvs(size=n)
Y = beta0 + beta1 * x + errors

df = pd.DataFrame({'x': x, 'y': Y})
model_data_specification = smf.ols("Y~x", data=df) 
fitted_model = model_data_specification.fit() 

df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='y', color='Data', 
                 trendline='ols', title='y vs. x')

# This is what `trendline='ols'` is
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

x_range = np.array([df['x'].min(), df['x'].max()])
y_line = beta0 + beta1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta0)+' + '+str(beta1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Add vertical lines for residuals
for i in range(len(df)):
    fig.add_scatter(x=[df['x'][i], df['x'][i]],
                    y=[fitted_model.fittedvalues[i], df['y'][i]],
                    mode='lines',
                    line=dict(color='red', dash='dash'),
                    showlegend=False)
    
# Add horizontal line at y-bar
fig.add_scatter(x=x_range, y=[df['y'].mean()]*2, mode='lines',
                line=dict(color='black', dash='dot'), name='y-bar')

fig.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

The "ordinary least squares" (OLS) method chooses the line that best fits the observed data by minimizing the sum of the squared residuals, which are the differences between the observed values and the predicted values on the fitted line.The purpose of squaring the residuals is to ensure that both positive and negative deviations are treated equally. Without squaring, positive and negative residuals could cancel each other out, misleadingly suggesting a perfect fit even if the data points deviate significantly from the model. By minimizing the sum of squared residuals, OLS ensures that the fitted line is as close as possible to all observed data points, while penalizing larger deviations more heavily than smaller ones, thereby producing a "best fit" line that captures the overall trend of the data.
In the provided visualization, the red dashed lines represent these residuals for each data point, showing the difference between the observed value and the fitted value on the trendline (in blue). The OLS method minimizes the length of these red lines squared, which ensures the fitted line accurately represents the linear trend in the observed data.

1.The numerator ((Y - fitted_model.fittedvalues)**2).sum() represents the sum of squared residuals (SSR), which quantifies the variation in the observed data (Y) that is not explained by the fitted model. It measures the discrepancy between the actual values (Y) and the predicted values (fitted_model.fittedvalues).
The denominator ((Y - Y.mean())**2).sum() represents the total sum of squares (TSS), which quantifies the total variation in the observed data (Y) from its mean. It measures how spread out the actual values are from the overall mean.
The entire fraction thus represents the unexplained proportion of variation by the model.
Therefore, 1 - (SSR / TSS) measures the proportion of the total variation in Y that is explained by the model. This proportion is also known as the coefficient of determination, or ùëÖ**2.
When ùëÖ**2 is close to 1, the model captures most of the variability in the outcome Y, implying a good fit.
When ùëÖ**2 is close to 0, the model explains very little of the variation, indicating a poor fit.


2.fitted_model.rsquared is a direct implementation of the R-squared value. It is essentially the same as the first expression and is used to evaluate the goodness of fit of the model. A higher R-squared value indicates that the model explains a greater proportion of the variance in the outcome Y. fitted_model.rsquared can be interpreted as a measure of the accuracy of the model, since it tells us how much of the variation in Y is captured by the predictor variables.

3.np.corrcoef(Y, fitted_model.fittedvalues)[0, 1] ** 2
The correlation coefficient (R) between two variables indicates the strength and direction of their linear relationship. In the context of linear regression, the correlation between the actual outcome values Y and the predicted values (fitted_model.fittedvalues) can be computed using np.corrcoef(Y, fitted_model.fittedvalues)[0, 1].

Squaring this value (** 2) gives the R-squared value, which measures the proportion of variation in Y that is explained by the linear relationship with the predicted values. This means that np.corrcoef(Y, fitted_model.fittedvalues)[0, 1] ** 2 is another way of calculating the R-squared value for the regression model, similar to the previous two methods.

np.corrcoef(Y, x)[0, 1] calculates the correlation coefficient between Y and x, which measures how strongly Y and x are linearly related.
Squaring this value (** 2) gives the R-squared value of the simple linear regression model, since in simple linear regression, the square of the correlation coefficient is equal to the proportion of the variance in Y explained by x.
Thus, np.corrcoef(Y, x)[0, 1] ** 2 captures the R-squared value for the simple linear regression model, showing the proportion of variance in Y explained by the predictor variable x. This indicates how well x explains the variability in Y.

Explanation of the Sampling Approaches
Permutation Test:

Purpose: To test if there‚Äôs a significant difference in means between the "short" and "long" groups.
Method: We shuffle the labels between "short" and "long" to create random groupings, assuming that there is no real difference between these groups.
Steps:
1.Randomly shuffle the kind column (e.g., old_faithful['kind']), creating a new column (kind_shuffled) with the labels randomly assigned.
2.Calculate the mean difference in duration between the shuffled groups.
3.Repeat this shuffling multiple times to create a distribution of mean differences under the null hypothesis (no difference between groups).
4.Compare the observed difference to this distribution to assess the likelihood of observing such a difference by chance.

Bootstrapping:
Purpose: To create a 95% confidence interval for the mean difference between "short" and "long" groups.
Method: We use sampling with replacement within each group separately to estimate the variability of the mean difference.
Steps:
1.For each group ("short" and "long"), sample with replacement to create a new dataset that represents a possible version of the population (bootstrapped samples).
2.Calculate the mean of each bootstrapped group, then compute the mean difference.
3.Repeat this process to obtain a distribution of bootstrapped mean differences.
4.Use the 2.5th and 97.5th percentiles of this distribution as the bounds for the 95% confidence interval.

Comparison to the Indicator Variable Based Model
The indicator variable approach likely uses regression to model the effect of being in the "short" vs. "long" group on the outcome, by including an indicator (binary) variable for group membership.  

Similarity:
Both the permutation and bootstrap methods, like the indicator variable model, aim to assess the difference between two groups.

Differences:
Permutation and Bootstrapping:
Directly simulate the distribution of differences under the null hypothesis or through resampling.
Are non-parametric and make fewer assumptions about the underlying data distribution.
The permutation test directly tests for differences under random assignment, while bootstrapping focuses on estimating confidence intervals for the mean difference.
Indicator Variable Approach:
Assumes a parametric relationship, often requiring assumptions like normality of residuals in regression.
Uses a binary variable to estimate the effect of group membership in a single model, rather than resampling.
Provides point estimates and confidence intervals based on model assumptions, rather than empirical distributions as in permutation and bootstrapping.
The indicator model tests categorical differences, using regression with a binary variable, while bootstrapping and simulation focus on continuous predictors (here, waiting duration).



````python
import pandas as pd
import numpy as np

# Load your Old Faithful dataset (assuming 'old_faithful' is already loaded)
# old_faithful = pd.read_csv('path_to_your_data.csv')  # Uncomment if loading from a file

# Step 1: Calculate the observed mean difference
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Step 2: Set up the permutation test
n_permutations = 1000  # Number of permutations
permuted_diffs = []

for _ in range(n_permutations):
    # Shuffle the 'kind' labels and assign them back to a new column
    old_faithful['kind_shuffled'] = old_faithful['kind'].sample(frac=1, replace=False).values
    # Calculate the mean difference for this permutation
    permuted_diff = old_faithful.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permuted_diffs.append(permuted_diff)

# Step 3: Calculate the p-value
permuted_diffs = np.array(permuted_diffs)
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_diff))

print("Observed Mean Difference:", observed_diff)
print("P-value:", p_value)
````

````python
import pandas as pd
import numpy as np

# Assuming old_faithful DataFrame is already loaded
# old_faithful = pd.read_csv('path_to_your_data.csv')  # Uncomment if loading from a file

# Number of bootstrap samples
n_bootstraps = 1000
bootstrapped_mean_differences = []

for _ in range(n_bootstraps):
    # Step 1: Create a bootstrap sample within each group
    bootstrapped_sample = old_faithful.groupby('kind').apply(
        lambda x: x.sample(n=len(x), replace=True)
    ).reset_index(drop=True)
    
    # Step 2: Calculate the mean difference for this bootstrap sample
    bootstrapped_diff = bootstrapped_sample.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_mean_differences.append(bootstrapped_diff)

# Step 3: Calculate the 95% confidence interval
bootstrapped_mean_differences = np.array(bootstrapped_mean_differences)
confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])

print("95% Bootstrap Confidence Interval for Mean Difference:", confidence_interval)
````
